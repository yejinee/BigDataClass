# 상관분석

```python
from sklearn import datasets
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
iris=datasets.load_iris()
```

- 값이 1에 가깝기 때문에 데이터 간에 강한 상관관계를 갖음 

```python
np.corrcoef(iris['data'][:,2],iris['data'][:,3])
```

```python
df=pd.DataFrame(iris['data'],columns=iris.feature_names)
```

```python
df['target']=iris['target']
```

```python
df.head(2)
```

```python
df.iloc[:,:4].corr() # 상관분석
```

## Visualization
- 상관계수에 대해서 heatmap으로 표현하면 효과적

```python
sns.heatmap(df.iloc[:,:4].corr(), annot=True, cmap='hot')
plt.ylim([4,0])
plt.xlim([4,0])
```
