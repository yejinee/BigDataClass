# 활성화 함수(Activation Function)
## ✔ 개념
- 신경망 회로에서 1개 node에 대한 입력값을 다음 node에 보낼지에 대해 결정하는 함수

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

## ✔ Step Function
- 0을 기준
    - 0보다 작으면, data 보내지 X
    - 0보다 크거나 같으면, data 보냄
- 실제로 사용X

```python
def step(x):
    result = x > 0.000001 # True , False
    return result.astype(np.int) # 정수로 변환
```

```python
x = np.arange(-10.0, 10.0, 0.1)
y = step(x)
plt.plot(x, y)
plt.show()
```

## ✔ Sigmoid Function
- Vanishing Gradient 문제 때문에 잘 쓰이지 X
- 0~1 사이의 결과값 가짐

### Vanishing Gradient
- 미분함수에서 x값이 커질수록 0에 수렴하는 형태 가짐

```python
def sigmoid(x):
    return 1.0 / ( 1.0 + np.exp(-x) )
```

```python
x = np.arange(-10., 10., 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

## ✔ Tanh Function
- Sigmoid와 유사한 형태 가짐
    - sigmoid에 비해 최적화 과정이 느려지는 문제 해결 가능 
- Vanishing Gradient 문제 발생

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-np.pi, np.pi, 60)
y = tanh(x)
plt.plot(x, y)
plt.show()
```

## ✔ ReLU Function
- 그래프 모양
    - x>0 : 기울기 1인 직선
    - x<0 : 함수값이 0됌
- Sigmoid, tanh와 비교했을 때 학습이 훨씬 빠름
- 연산비용 크지 X, 구현이 간단
- 일반적으로 제일 많이 쓰임
- 단점
    - x<0 인 값들에 대해 기울기가 0이기 때문에 뉴런이 죽을 수 있음
    - 보완한 버전으로 Leaky ReLU가 있음


```python
def relu(x):
    return np.maximum(x, 0)
```

```python
x = np.arange(-10., 10., 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```
