# Gradient Descent
## ✔ 개념
- 함수 최적화 기법 중 1개
- 미분의 개념을 최적화 문제에 적용 
    - 함수의 local minimum을 찾는 방법
    - 함수의 극대/극소 값을 구하기 위해 현재 위치에서 함수값의 변화가 큰 방향으로 이동

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = 10
learning_rate = 0.3
max_iter = 50
```

```python
# 로스 펑션
loss_func = lambda x: (x-3)**2 + 10
```

```python
# 그래디언트 펑션 : 손실함수의 1차 미분값
gradient = lambda x: 2*x-6
```

```python
list1 = []
list2 = []
```

```python
# 강하법
for i in range(max_iter):
    x = x - learning_rate * gradient(x)
    list1.append(x)
    list2.append(loss_func(x))
    print('손실함수값(', x, ')', loss_func(x))
```

```python
x1 = np.linspace(0.0, 10.0)
y1 = loss_func(x1)
fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.plot(list1, list2, '*')
```
