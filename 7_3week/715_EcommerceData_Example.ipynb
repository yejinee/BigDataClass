# E-commerce Data 이용한 Linear Regression
## 1. Import Module & Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('./ecommerce.csv')
```

## 2. Data의 속성 확인

```python
data.info()
```

```python
data.describe()
```

## 3. 그래프 그려보기

```python
sns.pairplot(data)
```

## 4. 불필요한 column 제거

```python
data.drop(['Email','Address','Avatar'], axis=1, inplace=True)
```

```python
data.head()
```

## 5. Train / Test Data 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('Yearly Amount Spent', axis=1)
```

```python
y=data['Yearly Amount Spent']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3, random_state=100)
```

## 6. Linear Regression Model 만들기

```python
import statsmodels.api as sm
```

```python
lm=sm.OLS(y_train, X_train).fit()
```

```python
lm.summary()
```

## 7. 예측 및 평가


```python
predictions =lm.predict(X_test)
```

```python
predictions
```

```python
y_test
```

```python
plt.figure(figsize=[10,10])
sns.scatterplot(y_test, predictions)
```

## Result
- graph가 선형적으로 나타난 것으로 보아 잘 예측되었음을 알 수 있음

## 8. 수치적으로 평가

```python
from sklearn import metrics
```

```python
print("Mse", metrics.mean_squared_error(y_test,predictions))
print("RMSE", np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

### Result
- 편차의 오차들이 오차가 21정도가 남 
- RMSE / MSE 가 0에 가까울수록 좋음
