# K-Means 간단 예제

## 1. 모듈 및 랜덤 데이터 생성

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, random_state=100)
```

```python
data
```

```python
data[0][:, 0]
data[0][:, 1]
```

```python
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1])
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(data[0])
```

```python
model.labels_
```

```python
data[1]
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue=data[1], palette='Set2')
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue=model.labels_, palette='Set2')
```

## 2. 다른 K값 사용해서 clustering

```python
model = KMeans(n_clusters=3)
model.fit(data[0])
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue=model.labels_, palette='Set2')
```

```python
model = KMeans(n_clusters=5)
model.fit(data[0])
plt.figure(figsize=(10, 8))
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue=model.labels_, palette='Set2')
```

## 3. 최적의 K값 찾기 : Elbow method

```python
model.inertia_
```

```python
distance = []
for i in range(2, 11):
    model = KMeans(n_clusters=i)
    model.fit(data[0])
    distance.append(model.inertia_)
```

```python
distance
```

```python
list(range(2, 11))
```

```python
sns.lineplot(x=list(range(2, 11)), y=distance)
```
